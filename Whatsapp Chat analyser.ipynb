```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with TE Computer _25.txt', 'r', encoding = 'utf-8')
```

```python
df = f.read()
```

```python
print(df)
```

```python
print(type(df))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
message = re.split(pattern,df)[1:]
```

```python
message
```

```python
dates = re.findall(pattern, df)
dates
```

```python
df = pd.DataFrame({'user_msg': message, 'msg_date': dates})
#convert to msg_date type
df['msg_date'] = pd.to_datetime(df['msg_date'], format = '%m/%d/%y, %H:%M - ')
df.rename(columns = {'msg_date': 'date'}, inplace = True)
```

```python
df
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_msg']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('Group Notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns = ['user_msg'], inplace = True)
```
